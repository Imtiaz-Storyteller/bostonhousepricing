# Preparing the dataset

```python
dataset=pd.DataFrame(boston.data, columns=boston.feature_names)
dataset
```

```python
dataset["Price"]=boston.target
```

```python
dataset
```

```python
#let's check the detailed information about the dataset
dataset.info()
```

```python
sns.pairplot(dataset)
```

```python
##Let's check using scatter and regression plot for better view.
for col in boston.feature_names:
    print(f"Plot between {col} and Price variable in Yaxis")
       
    sns.regplot(data=dataset,x=col,y="Price")
    plt.show()
```

```python
#Insights from the regplots, 
#1-When CRIM increase price decreases
#2-When LSTAT increase price decreases
#3-when PTRATIO increases price decreases
#4-when RM increase price also increases there is a positive corelationship between RM and price variables
```

```python
##Independent and dependent features
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
#Spliting dataset into train and test dataset
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.30,random_state=0)
```

```python
#Important step before training the model we need to scale the dataset. Because internally we use gradient desent mechanism
#all values to be in global mininum. Hence to bring the all the variable datapoints on same scale we perform the Standard scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(xtrain,ytrain)
```

```python
## to know the coefficient and intercept in the dataset
print("Coefficient:\n",regression.coef_, "\n\n", " Intercept:\n",regression.intercept_)
```

```python
#On which parameter the model has been trained

regression.get_params()
```

```python
##Prediction with test dataset
reg_predict=regression.predict(xtest)
```

```python
reg_predict
```

# Assumptions

```python
##Ploting the prediction
##Linear relationship between test and predict data points
plt.scatter(ytest,reg_predict);
```

```python
#calculating the residuals
residuals=ytest-reg_predict
residuals
```

```python
#plot this residuals
#Normally distributed
sns.displot(residuals, kind="kde");
```

```python
#When ever we plot the residuals it should be normally distributed, 
#in the above its almost normal distributed but there are some outliers present.
```

```python
#ploting the predict values and residuals
#Uniformly distributed
plt.scatter(reg_predict,residuals);
```

# Evaluating the model

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print("mean_absolute_error:", mean_absolute_error(ytest,reg_predict))
print("mean_squared_error:", mean_squared_error(ytest,reg_predict))
print("Sqrt_mean_squared_error:", np.sqrt(mean_squared_error(ytest,reg_predict)))
```

# R square and adjusted R square

formula

R square=1- SSR/SST

SSR sum of the square of residuals, SST sum of the square of total

```python
from sklearn.metrics import r2_score
score=r2_score(ytest,reg_predict)
print(score)
```

# Adjusted R2 = 1-[(1-R2)*(n-1)/(n-k-1)]

```python
1-(1-score)*(len(ytest)-1)/(len(ytest)-xtest.shape[1]-1)
```
